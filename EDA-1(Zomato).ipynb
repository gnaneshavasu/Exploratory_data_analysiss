Merging the two dataset reason is only require for country in second dataset based on country code in first dataset,so only add the country as column

```python
final=pd.merge(data,df_country,on="Country Code",how="left")
#final=pd.merge(data,df_country,on="Country Code",how="right")
```

```python
ratings=final.groupby(['Aggregate rating',"Rating color",'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

# Observation
1. When rating is between 4.5 to 4.9----->Excellent
2. when rating is between 4.0 to 4.4----->very Good
3. when rating is between 3.5 to 3.9----->Good
4. when rating is between 2.5 to 3.4----->Average
5. when rating is between 1.8 to 2.4----->poor



```python
ratings.head()
```

```python
plt.rcParams['figure.figsize']=(12 ,8) #its use to increase the plot size
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,width=2.0,hue="Rating text",palette=["blue",'red','orange','yellow','green',"green"]) 
```

Observation:
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4 
3. Good and very good ratings are between 4.0 to 4.9

sometimes if we have missing values in Aggregate rating. So analyse that maximum number of ratings are 2.5 to 3.4. So we take average between 2.5 to 3.4 after fill that values instead of missing values
